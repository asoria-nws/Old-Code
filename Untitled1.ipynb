```python
path = '/data/shared/ERA5_850mbV_eddyfilt_DJF_1by1deg.nc'
ds = xr.open_dataset(path)
ds
```

```python
pear = ds['time']
pear[0]
```

```python
date1=datetime.date(1979,12,1)
date2=datetime.date(2019,2,28)
times = pd.date_range(date1,date2)
time_da = xr.DataArray(times, [('time_da', times)])
time_da
```

```python
lat = ds['lat']
lon = ds['lon']
veddy = ds['v_eddy']
```

```python
veddy_array = xr.Dataset({
    'v_eddy': xr.DataArray(
                data   = veddy,
                dims   = ['time_da','lat','lon'],
                coords = {'time_da': time_da, 'lat': lat, 'lon': lon})})

veddy_array
```
